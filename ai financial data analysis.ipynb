## 7)Expenditure vs Revenue year-by-year?

```python
data.head()
```

```python
# important

# changed date to year 


spend=data.groupby(data['Date'].dt.year)['R&D_Spending_USD_Mn'].sum()
```

```python
revenue=data.groupby(data['Date'].dt.year)['AI_Revenue_USD_Mn'].sum()
```

```python
#plotting both 

plt.plot(spend.index,spend.values,color='orange')
plt.plot(revenue.index,revenue.values,color='green')
plt.title('spending and earning')
plt.xlabel('Year',fontsize=25)
plt.ylabel('USD in MILLION',fontsize=25)
plt.legend(['spending','revenue'])
```
